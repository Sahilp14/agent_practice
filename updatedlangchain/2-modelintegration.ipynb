### GROQ Model integration

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:qwen/qwen3-32b")
res = model.invoke("Hello, how are you?")
print(res.content)
```

```python
from langchain_groq import ChatGroq

groqmodel = ChatGroq(model_name="qwen/qwen3-32b")

res = groqmodel.invoke("why do parrots talk?")
print(res)
```

### Streaming

```python
for chunk in groqmodel.stream("Why do parrots have colorful feathers?"):
    print(chunk.text, end="", flush=True)
```

### Batch

```python
response = groqmodel.batch([
    "why do parrots have colorful feathers?",
    "what is the capital of France?",
    "what is the largest planet in the solar system?"],
    config = {
        "max_concurrency": 5,   # limit to 5 parallel calls
    }
)
for response in response:
    print(response)
```
